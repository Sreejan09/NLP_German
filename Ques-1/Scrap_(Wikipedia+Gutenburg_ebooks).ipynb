```python
def parse_articles(html):
    soup = BeautifulSoup(html, 'html.parser')
    articles = []

    # Adjust the tags and class names based on the website structure
    for article_tag in soup.find_all('article'):
        title_tag = article_tag.find('li',class_="mw-page-title-main")  # Adjust tag as necessary
        title = title_tag.get_text(separator="\n", strip=True)

        link = soup.find('link', rel="canonical")
        link = link['href']
        # print(link)
        articles.append({'title': title, 'link': link})

    return articles
```

```python
import requests
from bs4 import BeautifulSoup
import os
import re

# URL of the Portal:Geographie page
url = "https://de.wikipedia.org/wiki/Portal:Geographie"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the webpage content with BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find all <p> tags and extract links inside <a> tags
article_links = []
for p_tag in soup.find_all("p"):
    for a_tag in p_tag.find_all("a", href=True):
        link = a_tag['href']
        # Append full URL of the article link
        if link.startswith("/wiki/"):
            article_links.append("https://de.wikipedia.org" + link)

# Print the article links
for link in article_links:
    print(link)

# Optional: Save the links to a file
with open("geographie_article_links.txt", "w") as file:
    for link in article_links:
        file.write(link + "\n")

# Function to clean the title for a valid filename
def clean_filename(title):
    # Remove invalid characters (like /, \, :, *, ?, ", <, >, |)
    return re.sub(r'[\\/*?:"<>|]', "", title)

# Function to extract and save article text
def save_article_text(url, output_folder):
    article_response = requests.get(url)
    article_soup = BeautifulSoup(article_response.content, "html.parser")

    # Extract all text inside <p> tags
    article_text = ""
    for p in article_soup.find_all("p"):
        article_text += p.get_text() + "\n"

    # Get the article title for the filename
    title = article_soup.find("h1").get_text()
    filename = clean_filename(title) + ".txt"
    file_path = os.path.join(output_folder, filename)

    # Save the article text to a file
    with open(file_path, "w", encoding="utf-8-sig") as file:
        file.write(article_text)
    print(f"Saved article: {title}")

# Create folder to save articles
output_folder = "german_geographie_articles"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Extract and save text from each article link
for article_url in article_links:
    save_article_text(article_url, output_folder)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time
import os

# Base URL for German Wikipedia
base_url = "https://de.wikipedia.org"

# Set to track visited URLs across all batches
visited = set()

# Counter for the batch files
batch_counter = 1

# Function to scrape 500 articles in a batch and save to a file
def scrape_batch(start_url, batch_size=500):
    global batch_counter
    stack = [start_url]  # Stack to hold URLs to scrape
    scraped_count = 0  # Track number of articles scraped in this batch
    batch_file = f"wikipedia_batch_{batch_counter}.txt"

    # Open a new file for the current batch
    with open(batch_file, 'w', encoding='utf-8') as file:

        while stack and scraped_count < batch_size:
            url = stack.pop()  # Get the next URL to scrape

            if url in visited:
                continue  # Skip if already visited

            # Mark the URL as visited
            visited.add(url)

            # Fetch page content
            try:
                response = requests.get(url)
                response.raise_for_status()  # Check for errors
                soup = BeautifulSoup(response.text, 'html.parser')

                # Extract the page title and text content
                title = soup.find('h1').text.strip()
                paragraphs = soup.find_all('p')  # Grabbing all <p> tags for text content

                # Write content to the file
                file.write(f"\n\nTitle: {title}\n")
                file.write(f"URL: {url}\n\n")
                for para in paragraphs:
                    text = para.get_text().strip()
                    file.write(f"{text}\n")

                print(f"Scraped: {title} ({url})")

                # Increment the scraped count
                scraped_count += 1

                # Find all <a> tags and extract href links
                for a_tag in soup.find_all('a', href=True):
                    link = a_tag['href']

                    # Ensure the link is within Wikipedia and not an external site
                    if link.startswith("/wiki/") and not ":" in link:  # Skip special pages like /wiki/Help:
                        full_link = urljoin(base_url, link)

                        # Only add new links to the stack
                        if full_link not in visited:
                            stack.append(full_link)

            except requests.RequestException as e:
                print(f"Error fetching {url}: {e}")

    # Close the file after the batch is scraped
    print(f"Batch {batch_counter} saved: {batch_file}")
    batch_counter += 1
    return stack  # Return the current stack to continue scraping from where we left off

# Start scraping
start_url = urljoin(base_url, "/wiki/Wikipedia:Hauptseite")

# Set batch size
batch_size = 500

# Initial call to scrape the first batch
stack = scrape_batch(start_url, batch_size)

# Continue scraping the next batches
while stack:
    stack = scrape_batch(stack.pop(), batch_size)
```

```python
import requests
from bs4 import BeautifulSoup

# Base URL for the search results
base_url = "https://textgridrep.org/search?query=&order=relevance&limit=20&mode=list&filter=edition.language:deu&filter=format%3Atext%2Fxml&page="

# Function to scrape one page
def scrape_page(page_num):
    page_url = base_url + str(page_num)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all article links
    articles = soup.find_all('a', href=True)

    # Initialize text for all articles on the page
    page_text = ""

    # Loop through each article link
    for article in articles:
        article_url = "https://textgridrep.org" + article['href']

        # Scrape each article page
        article_response = requests.get(article_url)
        article_soup = BeautifulSoup(article_response.text, 'html.parser')

        # Extract the title and content of the article
        title = article_soup.find('title').text.strip()
        article_text = article_soup.get_text()

        # Append the title and content to the page_text
        page_text += f"Title: {title}\n\n{article_text}\n\n{'-'*50}\n\n"

    return page_text

# Scrape multiple pages and store all articles in a single file
def scrape_multiple_pages(start_page, end_page):
    full_text = ""

    for page_num in range(start_page, end_page + 1):
        full_text += scrape_page(page_num)
        print(f"Scraped page {page_num}")

    # Save the full text to a file
    with open("all_pages_articles.txt", "w", encoding="utf-8") as file:
        file.write(full_text)

# Scrape from page 1 to page 3 (you can adjust the range)
scrape_multiple_pages(1, 200)

print("All pages scraped and saved in 'all_pages_articles.txt'.")
```
